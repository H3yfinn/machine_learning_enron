```python
def targetFeatureSplit( data ):
    """ 
        given a numpy array like the one returned from
        featureFormat, separate out the first feature
        and put it into its own list (this should be the 
        quantity you want to predict)

        return targets and features as separate lists

        (sklearn can generally handle both lists and numpy arrays as 
        input formats when training/predicting)
    """

    target = []
    features = []
    for item in data:
        target.append( item[0] )
        features.append( item[1:] )

    return target, features

labels, features = targetFeatureSplit(log_df)

mmscaler = preprocessing.MinMaxScaler()
features = mmscaler.fit_transform(features)
```

```python
#MACHINE LEARNING TIME!!
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA
from sklearn import cross_validation
from sklearn.grid_search import GridSearchCV
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn import ensemble
from sklearn.pipeline import Pipeline
from sklearn.cross_validation import train_test_split
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn import linear_model
from time import time
sep = '##############################################################################################'
sep2 = '++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++'
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=0.33, random_state=40)
target_names = ['NonPoi', 'Poi']
```

I'm going to test 6 different supervised learning algorithms with the aim of finding the most accurate. I'll be testing them in a pipeline twice each, once using the SelectKbest feature selctor, and then again using PCA instead to identify/merge the best set of features. 
It's important that we validate these models by checking their accuracy, recall, precision and F1 scores. This will be done with the test set's. Cross validation will be used with gridsearchCV on the training sets to get the best hyperparameters. Lastly we will record their run time to show how each model differs in it's efficiency.
May the best model win!


```python
def test_pipe(classifier, selector, params, classifier_name):
    #easy function designed to decrease workload by a lot
    t = time()
    if selector == 'PCA':
        paramaters = {'SEL__n_components': [13, 12, 11, 10, 9, 8, 7]}
        paramaters.update(params)
        selector_ = PCA()
    elif selector == 'Kbest':
        paramaters = {'SEL__k':[7,8,9,10,11,12,13,14,15]}
        paramaters.update(params)
        selector_ = SelectKBest()
    else:
        raise Exception('Something went wrong, bugfixing time!') 
    
    cv = cross_validation.StratifiedShuffleSplit(labels_train, n_iter = 10, random_state = 42)
    pipe = Pipeline(steps=[('SEL', selector_), ('CLF', classifier)])
    gs = GridSearchCV(pipe, param_grid = paramaters,
                              scoring = 'f1', cv =cv)

    gs.fit(features_train, labels_train)
    pred = gs.predict(features_test)
    print 'classifier:', classifier_name
    print 'Selector:', selector
    print classification_report(labels_test, pred, target_names=target_names)
    print 'accuracy=', accuracy_score(labels_test, pred)
    print 'paramaters:', gs.best_params_
    print "training time:", round(time()-t, 3), "s"
```

I found that Logistic Regression was the best classifier algorithm to use with Principle component analysis. 
The most important evalutation metrics are precision and recall for Poi classifications. They are at 0.31 and 0.71 for precison and recall respectively. The accuracy score at 0.729 is less important as it is easily swayed by the model's propensity to guess 'NonPoi'. For example, if the model was to guess that all the data points in the test set were Non-poi's then we would already have an accuracy of 86%. Yet it would be more desireable for the model to run with a lower accuracy but greater propensity for picking poi's to raise the Recall score for Poi's. 
I believe the recall score is the most important because the aim of running this algorithm is to uncover potential POI's for further investigation. As a result we'd rather have a few false positives than false negatives for Poi's. -false positives are those that are guessed to be poi's but arent and false negatives are those who are assumed to be non-poi's but are poi's. Thus, higher recall is preffered over precision.
